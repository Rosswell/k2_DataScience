```python
% matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import glob
```

```python
url_template = 'http://web.mta.info/developers/data/nyct/turnstile/turnstile_%s.txt'
for date in ['170603', '170527', '170520', '170513']:
    url = url_template % date
    !wget {url}
```

```python
file_list = glob.glob("turnstile_*.txt")
for file in file_list:
    if file_list.index(file) == 0:
        MTA_data = pd.read_csv(file)
    else:
        MTA_data = pd.concat([MTA_data, pd.read_csv(file)], ignore_index=True)
MTA_data.info()
```

```python
MTA_data.head()
```

```python
MTA_data.DATE = pd.to_datetime(MTA_data.DATE)
```

```python
MTA_data['TURNSTILE'] = MTA_data['C/A'] + " " + MTA_data['UNIT'] + " " + MTA_data['SCP']
```

```python
MTA_data.TIME = pd.to_datetime(MTA_data.TIME, format='%H:%M:%S')
MTA_data.TIME = MTA_data.TIME.apply(lambda x: datetime.time(x))
```

```python
MTA_data.head()
```

```python
MTA_data.drop(['C/A', 'UNIT', 'SCP', 'LINENAME', 'DESC', 'DIVISION'], axis=1, inplace=True)
MTA_data.columns = ['STATION', 'DATE', 'TIME', 'ENTRIES', 'EXITS', 'TURNSTILE']
```

```python
from collections import Counter
foottraffic_dict = Counter()
for station in MTA_data.STATION.unique():
    station_data = MTA_data[MTA_data.STATION == station]

    for turnstile in station_data.TURNSTILE.unique():
        total_foottraffic = 0
        turnstile_data = station_data[station_data.TURNSTILE == turnstile]

        total_entries = turnstile_data.ENTRIES.max() - turnstile_data.ENTRIES.min()
        if total_entries < 1000000:
            total_foottraffic += total_entries

        total_exits = turnstile_data.EXITS.max() - turnstile_data.EXITS.min()
        if total_exits < 1000000:
            total_foottraffic += total_exits

        foottraffic_dict[station] += total_foottraffic
```

```python
from operator import itemgetter
foottraffic_list = []
for k, v in foottraffic_dict.items():
    foottraffic_list.append((k, v))
```

Finding the top 10 highest foot traffic stations

```python
foottraffic_sorted = sorted(foottraffic_list, key=itemgetter(1), reverse=True)[:10]
```

The stations that have the highest foot traffic are

```python
x_array = np.arange(10)
x_obs = [i[0] for i in foottraffic_sorted]
y_values = [i[1] for i in foottraffic_sorted]
plt.figure(figsize=(20,10))
plt.bar(x_array, y_values)
plt.xticks(x_array, x_obs)
plt.show()
```

Looking at the traffic on a daily basis

```python
days = ['Saturday', 'Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']

day_series = pd.Series()
weekday_list = []
days_dict = {}
for day in MTA_data.DATE.unique():
    
    day_index = (day - MTA_data.DATE.unique()[0]).astype('timedelta64[D]').astype('int')%7
    
    days_dict[day.astype('datetime64[D]')] = days[day_index]

for x in MTA_data.DATE:
    x = np.datetime64(x).astype('datetime64[D]')
    weekday_list.append(days_dict[x])

MTA_data['WEEKDAY'] = weekday_list
```

```python
from collections import defaultdict
day_dict = defaultdict(list)
for station in x_obs:
    station_data = MTA_data[MTA_data.STATION == station]

    for day in days:
        day_data = station_data[station_data.WEEKDAY == day]
        
        total_foottraffic = 0
        for turnstile in day_data.TURNSTILE.unique():
            turnstile_data = day_data[day_data.TURNSTILE == turnstile]

            total_entries = turnstile_data.ENTRIES.max() - turnstile_data.ENTRIES.min()
            if total_entries < 1000000:
                total_foottraffic += total_entries

            total_exits = turnstile_data.EXITS.max() - turnstile_data.EXITS.min()
            if total_exits < 1000000:
                total_foottraffic += total_exits

        day_dict[station].append([day, total_foottraffic])
```

```python
import seaborn as sns
days_for_df = {'days': ['Saturday', 'Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']}
plot_df = pd.DataFrame(days_for_df, columns=['days'])
for station in day_dict:
    y_vals = []
    for day_traffic in day_dict[station]:
        y_vals.append(day_traffic[1])
    plot_df[station] = y_vals
    temp_sum = plot_df[station].sum()
    plot_df[station] = plot_df[station] / temp_sum
plot_df.plot(figsize=(15,10))
x_array = np.arange(len(days))
plt.title('Foot Traffic Distribution by Weekday', size=20)
plt.ylabel('Foot Traffic Ratio')
plt.xlabel('Weekday')
plt.xticks(x_array, days)
plt.show()
```

```python
def time_in_range(start_end, x):
    """Return true if x is in the range [start, end]"""
    start, end = start_end
    if start <= end:
        return start <= x <= end
    else:
        return start <= x or x <= end
```

```python
time_list = [[datetime.time(0,0,0), datetime.time(4,0,0)], 
             [datetime.time(4,0,0), datetime.time(8,0,0)],
             [datetime.time(8,0,0), datetime.time(12,0,0)], 
             [datetime.time(12,0,0), datetime.time(16,0,0)],
             [datetime.time(16,0,0), datetime.time(20,0,0)],
             [datetime.time(20,0,0), datetime.time(0,0,0)]]

time_dict = defaultdict(list)
for station in x_obs:
    station_data = MTA_data[MTA_data.STATION == station]

    for time in time_list:
        time_data = station_data[station_data['TIME'].apply(lambda x: time_in_range(time, x))]
        
        total_foottraffic = 0
        for turnstile in time_data.TURNSTILE.unique():
            turnstile_data = time_data[time_data.TURNSTILE == turnstile]
            
            total_entries = turnstile_data.ENTRIES.max() - turnstile_data.ENTRIES.min()
            if total_entries < 1000000:
                total_foottraffic += total_entries

            total_exits = turnstile_data.EXITS.max() - turnstile_data.EXITS.min()
            if total_exits < 1000000:
                total_foottraffic += total_exits

        time_dict[station].append([time, total_foottraffic])
```

```python
time_range_list = ['00h - 04h', '04h - 08h', '08h - 12h', '12h - 16h', '16h - 20h', '20h - 00h']
times_for_df = {'times': time_range_list}
time_plot_df = pd.DataFrame(times_for_df, columns=['times'])
for station in day_dict:
    y_vals = []
    for time_traffic in time_dict[station]:
        y_vals.append(time_traffic[1])
    time_plot_df[station] = y_vals
    temp_sum = time_plot_df[station].sum()
    time_plot_df[station] = time_plot_df[station] / temp_sum
time_plot_df.plot(figsize=(15,10))

x_array = np.arange(len(time_range_list))
plt.title('Foot Traffic Distribution by Time Range', size=20)
plt.ylabel('Foot Traffic Ratio')
plt.xlabel('Time')
plt.xticks(x_array, time_range_list)
plt.show()
```
