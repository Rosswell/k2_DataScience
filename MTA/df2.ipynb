```python
import pandas as pd
import numpy as np
import csv
import glob
from collections import defaultdict
```

```python
def dict_create(file):
    MTA_dict = defaultdict(list)
    
    with open(file) as f:
        file_reader = csv.reader(f)
        
        for row, col in enumerate(file_reader):
            #skipping headers
            if row == 0:
                continue
            #splitting the rows in 2 for keys and values
            dict_key = tuple(col[:4])
            dict_values = col[4:]
            # appending multiple rows to a single key
            MTA_dict[dict_key].append(dict_values)
    return MTA_dict
```

```python
MTA_data = [dict_create(file) for file in glob.glob('*.txt')]   
```

## Exercise 2
- Let's turn this into a time series.

 For each key (basically the control area, unit, device address and station of a specific turnstile), have a list again, but let the list be comprised of just the point in time and the cumulative count of entries.

This basically means keeping only the date, time, and entries fields in each list. You can convert the date and time into datetime objects -- That is a python class that represents a point in time. You can combine the date and time fields into a string and use the [dateutil](https://dateutil.readthedocs.io/en/stable/) module to convert it into a datetime object.

Your new dict should look something like
 
    {    ('A002','R051','02-00-00','LEXINGTON AVE'):    
             [
                [datetime.datetime(2013, 3, 2, 3, 0), 3788],
                [datetime.datetime(2013, 3, 2, 7, 0), 2585],
                [datetime.datetime(2013, 3, 2, 12, 0), 10653],
                [datetime.datetime(2013, 3, 2, 17, 0), 11016],
                [datetime.datetime(2013, 3, 2, 23, 0), 10666],
                [datetime.datetime(2013, 3, 3, 3, 0), 10814],
                [datetime.datetime(2013, 3, 3, 7, 0), 10229],
                ...
              ],
     ....
     }



```python
from dateutil.parser import parse
```

```python
def dt_parser(full_dict):
    time_series_dict = defaultdict(list)
    
    for week in full_dict:
        for key, value in week.items():
            date = []
            for entry in value:
                # replacing the dates slashes with dashes and parsing as a datetime object
                datetime_object = parse(' '.join([entry[2].replace('/', '-'), entry[3]]))
                entry_count = int(entry[5])
                # appending datetime object and count to the list, then adding the list to dict 
                # values
                date.append([datetime_object, entry_count])
            time_series_dict[key].append(date)
    return time_series_dict
```

```python
MTA_time_series = dt_parser(MTA_data)
```

```python
# all the weeks are in different lists, need to combine them
def week_combine(MTA_data):
    combined_data = defaultdict(list)
    
    for week, time_series in MTA_data.items():
        for hour in time_series:
            combined_data[week] += hour
    return combined_data
```

```python
def weekly_series(MTA_dict):
    daily_intake_dict = {}
    
    for row, (turnstile, time_series) in enumerate(MTA_dict.items()):
        print("processing turnstile", turnstile)
        daily_intake_dict[turnstile] = daily_calculation(time_series)
        
    return daily_intake_dict
```

```python
from itertools import groupby

def count_within_normal_bounds(count):
    # making sure the difference between days is positive or 0
    if count is None:
        return True
    else:
        return 10000 > count >= 0

def daily_calculation(time_series):
    daily_time_series = []
    
    def day_of_timestamp(entry):
        # returning just the date of the datetime object
        timestamp, total_entries = entry
        return timestamp.date()
    
    count_on_previous_day = None
    for day, entries_on_this_day in groupby(time_series, key=day_of_timestamp):
        total_entry_count_on_day = max([count for time, count in entries_on_this_day])
        
        # calculating the difference between subsequent days
        if count_on_previous_day is None:
            daily_entries = None
        else:
            daily_entries = total_entry_count_on_day - count_on_previous_day
        
        count_on_previous_day = total_entry_count_on_day
        
        # error message for irregular values
        if count_within_normal_bounds(daily_entries):
            daily_time_series.append((day, daily_entries))
        else:
            print("Weird numbers found on day %s: %s" % (day, daily_entries))
            daily_time_series.append((day, None))
            
    return daily_time_series
```

## Exercise 5
- So far we've been operating on a single turnstile level, let's combine turnstiles in the same ControlArea/Unit/Station combo. There are some ControlArea/Unit/Station groups that have a single turnstile, but most have multiple turnstilea-- same value for the C/A, UNIT and STATION columns, different values for the SCP column.

We want to combine the numbers together -- for each ControlArea/UNIT/STATION combo, for each day, add the counts from each turnstile belonging to that combo.


```python
def combine_turnstiles(daily_time_series):
    # splitting the keys into their elements, creating new keys
    daily_turnstile_grouping = defaultdict(list)
    for k, v in daily_time_series.items():
        CA, UNIT, SCP, STATION = k
        daily_turnstile_grouping[(CA, UNIT, STATION)].append(v)
    return daily_turnstile_grouping
```

```python
daily_turnstile_grouping = combine_turnstiles(daily_time_series)
```

```python
pprint(list(daily_turnstile_grouping.items())[:3])
```

```python
from collections import Counter

def adding_counts(daily_turnstile_grouping):
    # lots of for loops to get within the data, add counts for a common key with Counter dict
    daily_total_counts = {}
    
    for booth, daily_serieses in daily_turnstile_grouping.items():
        daily_cumulative = Counter()
        for turnstile in daily_serieses:
            for day, count in turnstile:
                if count is not None:
                    daily_cumulative[day] += count
        daily_total_counts[booth] = sorted(daily_cumulative.items())
    
    return daily_total_counts
```

```python
daily_total_counts = adding_counts(daily_turnstile_grouping)
pprint(daily_total_counts[('A002', 'R051', '59 ST')])
```

## Exercise 6
- Similarly, combine everything in each station, and come up with a time series of `[(date1, count1),(date2,count2),...]` type of time series for each STATION, by adding up all the turnstiles in a station.

```python
def combine_stations(daily_total_counts):
    # splitting the keys into their respective categories, creating new keys
    daily_station_grouping = defaultdict(list)
    for k, v in daily_total_counts.items():
        CA, UNIT, STATION = k
        daily_station_grouping[STATION].append(v)
    return daily_station_grouping
```

```python
daily_station_grouping = combine_stations(daily_total_counts)
pprint(daily_station_grouping['59 ST'])
```

```python
def adding_station_counts(daily_station_grouping):
    # lots of for loops to get within the data and sum the counts for the same day
    daily_station_counts = {}
    
    for station, daily_serieses in daily_station_grouping.items(): 
        daily_cumulative = Counter()
        for turnstile in daily_serieses:
            for day, count in turnstile:
                if count is not None:
                    daily_cumulative[day] += count
        daily_station_counts[station] = sorted(daily_cumulative.items())
    
    return daily_station_counts
```

```python
daily_station_totals = adding_station_counts(daily_station_grouping)
pprint(daily_station_totals['59 ST'])
```

## Exercise 8
- Make one list of counts for **one** week for one station. Monday's count, Tuesday's count, etc. so it's a list of 7 counts.
Make the same list for another week, and another week, and another week.
`plt.plot(week_count_list)` for every `week_count_list` you created this way. You should get a rainbow plot of weekly commute numbers on top of each other.



```python
def plot_weeks(daily_station_totals, station):
    # establishing the lists for graphing
    weekdays = ['Sunday','Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
    station_datecounts = daily_station_totals[station]
    week_count_list = []
    plt.figure(figsize=(15,5))
    
    for date, count in station_datecounts:
        week_count_list.append(count)
        # plots and then resets the week_count when the count length is 7 or it's the last week
        if len(week_count_list) % 7 == 0 or (date, count) == station_datecounts[-1]:
            plt.plot(week_count_list)
            week_count_list = []
    
    # establishing the x axis 
    x = np.array(np.arange(7))
    plt.xlabel("Days")
    plt.ylabel("Passengers traveling through the station per day")
    # jupyter notebooks reacts weirdly to fstrings
    plt.title(f"Passengers traveling through {station}")
    plt.xticks(x, weekdays)
    plt.show()
```

```python
plot_weeks(daily_station_totals, '59 ST')
```

```python
def total_ridership(daily_station_totals):
    # summing counts over individual stations
    total_ridership_count = Counter()
    for station, datecount in daily_station_totals.items():
        for date, count in datecount:
             total_ridership_count[station] += count
    return total_ridership_count
```

```python
total_ridership_dict = total_ridership(daily_station_totals)
pprint(list(total_ridership_dict.items()))
```
