```python
features_df.var()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
scaled_features = pd.DataFrame(scale(features_df), index=features_df.index, columns=features_df.columns)
```

```python
pca_loadings = pd.DataFrame(PCA().fit(scaled_features).components_.T, index=scaled_features.columns, columns=['V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7'])
pca_loadings
```

```python
pca = PCA()
df_plot = pd.DataFrame(pca.fit_transform(scaled_features),
                       columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 'PC7'], index=scaled_features.index)
df_plot.head()
```

```python
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```

```python
plt.figure(figsize=(7,5))
plt.plot([1,2,3,4,5,6,7], pca.explained_variance_ratio_, '-o')
plt.ylabel('Proportion of Variance Explained')
plt.xlabel('Principal Component')
plt.xlim(0.75,7.25)
plt.ylim(0,0.35)
plt.xticks([1,2,3,4,5,6,7])
plt.show()
```

Clearly Componenents 6 and 7 do not contribute to the data Variance very much, and can be discarded.

```python
plt.figure(figsize=(7,5))
plt.plot([1,2,3,4,5,6,7], np.cumsum(pca.explained_variance_ratio_), '-s')
plt.ylabel('Proportion of Variance Explained')
plt.xlabel('Principal Component')
plt.xlim(0.75,7.25)
plt.ylim(0.2,1.05)
plt.xticks([1,2,3,4,5,6,7])
plt.show()
```

```python
pd.DataFrame([df_plot.iloc[:,:7].std(axis=0, ddof=0).as_matrix(),
              pca.explained_variance_ratio_[:7],
              np.cumsum(pca.explained_variance_ratio_[:7])],
             index=['Standard Deviation', 'Proportion of Variance', 'Cumulative Proportion'],
             columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 'PC7'])
```

```python
fig , (ax1,ax2) = plt.subplots(1,2, figsize=(15,5))

# Left plot
ax1.plot(pca.explained_variance_ratio_, '-o')
ax1.set_ylabel('Proportion of Variance Explained')

# Right plot
ax2.plot(np.cumsum(pca.explained_variance_ratio_), '-ro')
ax2.set_ylabel('Cumulative Proportion of Variance Explained')

for ax in fig.axes:
    ax.set_xlabel('Principal Component')
    ax.set_xlim(-1,6.5)   
```

```python
plt.figure(figsize=(15,5))
plt.bar(range(16), pca.explained_variance_ratio_, color='m', alpha=0.5, label='Proportion of Variance Explained')
plt.step(range(16), np.cumsum(pca.explained_variance_ratio_), label='Cumulative Proportion of Variance Explained')
plt.plot(range(16), [0.95] * 16, color='k', label='95% Variance')
plt.xlabel('Principal Component')
plt.legend()
```
