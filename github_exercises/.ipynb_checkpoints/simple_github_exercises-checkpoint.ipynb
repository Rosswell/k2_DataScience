## 6. Create 2 tables: `Company` and `Job`. Pay special attention on primary and foreign keys as well as unique and non-nullable columns. Make sure you store `created_at` as the equivalent of Python `datetime` and not `string`.


```python
from datetime import datetime
# creating a list of the companies for the company table
company_list = []
number_of_companies = len(posting_data)
for index in range(number_of_companies):
    company_list.append(posting_data[index]['company'])
    
    # if the created_at column entry is a string rather than datetime, convert to datetime
    if isinstance(posting_data[index]['created_at'], str):
        posting_data[index]['created_at'] = datetime.strptime(posting_data[index]['created_at'], '%a %b %d %H:%M:%S %Z %Y')
```

```python
# Creating the Company table
c.execute("""CREATE TABLE IF NOT EXISTS Company (
Company_name text NOT NULL,
id text NOT NULL PRIMARY KEY)""")

# creating the Job table 
c.execute("""CREATE TABLE IF NOT EXISTS Job (
id text NOT NULL PRIMARY KEY,
created_at date_time NOT NULL,
title text NOT NULL,
location text,
type text, 
description text NOT NULL, 
how_to_apply text NOT NULL,
company text NOT NULL,
company_url text NOT NULL,
company_logo text NOT NULL,
url text NOT NULL
)""")

# Committing changes and closing the connection to the database file
conn.commit()
```

## 7. Store the data you got from the API request in the database. Make sure you create a transaction (i.e., if an error occurs while saving a job's location data the database should rollback on all statements regarding that job).

```python
company_count = 0
# inserts the id into the company_list list - forgot to add it when creating the list.
# need the id to create a primary key in the company list; company is not unique
for insertion in range(1, (len(company_list))*2, 2):
    company_list.insert(insertion, posting_data[company_count]['id'])
    company_count += 1
```

```python
# write data to company table

# creating a series of question marks and parentheses for multi-adding rows to the company table
placeholder = ('(?, ?), '*(len(company_list)//2))[:-2]

# string of question marks the length of the number of columns
company_table_query = 'INSERT INTO Company (Company_name, id) VALUES %s' % placeholder
c.execute(company_table_query, company_list)
conn.commit()
```

```python
# list of the columns to reference within the job table when adding rows
column_list = ', '.join(list(job_posting.keys()))
```

```python
value_list = []
# creates a list of the values of the various keys of the job postings, puts them in value_list
for posting in range(number_of_companies):
    temp_list = [str(posting_data[posting][key]) for key in list(job_posting.keys())]
    value_list.append(temp_list)
    
# function to flatten out list of lists
flatten = lambda l: [item for sublist in l for item in sublist]

#flattening the list for use in the query to add values to the Job table
value_list = flatten(value_list)
```

```python
# SQLite requires this type of substitution to mimic sequences. see DBAPI paramstyle (annoying)
placeholder = ('(?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?), '*(len(value_list)//11))[:-2]

query = 'INSERT INTO Job (%s) VALUES %s' % (column_list, placeholder)
c.execute(query, value_list)
conn.commit()
```

## 9. Create queries the following queries:
 * Find out how many jobs have the phrase `Data Science` and how many have the word `Python` in the description.
 * Find out how many were created in the past 7 days.
 * Find out how many jobs are located in `California` and which companies those are with.

**The output from our queries should be similar to this:**

There are 27 jobs with the 'Data Science' term, and 40% of them have 'Python' in the description.

22% of them have been created in the past 7 days.

2 of those are in California. The companies with those jobs are: ['Carbon Black, Inc.', 'Kaggle']

```python
# job description containing 'data science' query
query0 = "SELECT COUNT(*) FROM Job WHERE Description LIKE '%Data Science%'"

# job description containing 'python' query, as a percentage of the total rows query
query1 = "SELECT CAST((SELECT COUNT(*) FROM Job WHERE Description LIKE '%Python%') AS float)/(SELECT COUNT(*) FROM Job) * 100  AS Percentage"
c.execute(query0)
data_science = c.fetchone()[0]
c.execute(query1)
python = c.fetchone()[0]
print("There are {} jobs with the 'Data Science' term, and {}% of them have 'Python' in the description.".format(data_science, python))
```

```python
# created_at is more recent than 7 days ago as a percentage of the total rows present query
query2 = "SELECT CAST((SELECT COUNT(*) FROM Job WHERE date(created_at) >= date('now','-7 days')) AS float)/(SELECT COUNT(*) FROM Job) * 100  AS Percentage"# % week_ago
c.execute(query2)
print("{}% of them have been created in the past 7 days.".format(c.fetchone()[0]))
```

```python
# count of locations containing 'california'
query3 = "SELECT COUNT(*) FROM Job WHERE location LIKE '%California%'"
c.execute(query3)
cali_number = c.fetchone()[0]

# table join on id to get the company name of the posting id that contains 'california'
query4 = """SELECT Company_name FROM Company 
INNER JOIN Job
ON Job.id = Company.id
WHERE location LIKE '%California%'"""
c.execute(query4)
Company_name = c.fetchone()[0]
print("{} of those are in California. The companie(s) with those job(s) are: {}".format(cali_number, Company_name))
```
